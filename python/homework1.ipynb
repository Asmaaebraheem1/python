```python
#create a small program where users can add,delete ,and view contacts in a text file each contact will have a name and add phone number
import os

def load_contacts(filename):
    contacts = {}
    if os.path.exists(filename):
        with open(filename, 'r') as file:
            for line in file:
                name, phone = line.strip().split(',')
                contacts[name] = phone
    return contacts

def save_contacts(contacts, filename):
    with open(filename, 'w') as file:
        for name, phone in contacts.items():
            file.write(f"{name},{phone}\n")

def add_contact(contacts, name, phone):
    contacts[name] = phone
    print(f"Contact '{name}' added.")

def delete_contact(contacts, name):
    if name in contacts:
        del contacts[name]
        print(f"Contact '{name}' deleted.")
    else:
        print(f"Contact '{name}' not found.")

def view_contacts(contacts):
    if contacts:
        print("Contacts:")
        for name, phone in contacts.items():
            print(f"{name}: {phone}")
    else:
        print("No contacts found.")

def main():
    filename = "contacts.txt"
    contacts = load_contacts(filename)

    while True:
        print("\n1. Add contact")
        print("2. Delete contact")
        print("3. View contacts")
        print("4. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            name = input("Enter name: ")
            phone = input("Enter phone number: ")
            add_contact(contacts, name, phone)
            save_contacts(contacts, filename)
        elif choice == '2':
            name = input("Enter name to delete: ")
            delete_contact(contacts, name)
            save_contacts(contacts, filename)
        elif choice == '3':
            view_contacts(contacts)
        elif choice == '4':
            print("Exiting.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
#second question
"""write a program that allows users to add tasks to a a to do list ,mark them as completed ,and save the do to list to a file 
each task could be saved in the format "task,completed" in the file (e.g,"wishes,false")
allow reading ,adding and updating tasks"""
import os

def load_tasks(filename):
    """Load tasks from the text file."""
    tasks = {}
    if os.path.exists(filename):
        with open(filename, 'r') as file:
            for line in file:
                task, completed = line.strip().split(',')
                tasks[task] = completed.lower() == 'true'
    return tasks

def save_tasks(tasks, filename):
    """Save tasks to the text file."""
    with open(filename, 'w') as file:
        for task, completed in tasks.items():
            file.write(f"{task},{completed}\n")

def add_task(tasks, task):
    """Add a new task."""
    if task in tasks:
        print(f"Task '{task}' already exists.")
    else:
        tasks[task] = False
        print(f"Task '{task}' added.")

def mark_task_completed(tasks, task):
    """Mark a task as completed."""
    if task in tasks:
        tasks[task] = True
        print(f"Task '{task}' marked as completed.")
    else:
        print(f"Task '{task}' not found.")

def view_tasks(tasks):
    """View all tasks."""
    if tasks:
        print("To-Do List:")
        for task, completed in tasks.items():
            status = "True" if completed else "False"
            print(f"{task}: {status}")
    else:
        print("No tasks found.")

def main():
    filename = "tasks.txt"
    tasks = load_tasks(filename)

    while True:
        print("\n1. Add task")
        print("2. Mark task as completed")
        print("3. View tasks")
        print("4. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            task = input("Enter task: ")
            add_task(tasks, task)
            save_tasks(tasks, filename)
        elif choice == '2':
            task = input("Enter task to mark as completed: ")
            mark_task_completed(tasks, task)
            save_tasks(tasks, filename)
        elif choice == '3':
            view_tasks(tasks)
        elif choice == '4':
            print("Exiting.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
